# Ridge Regression

```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
%matplotlib inline
```

### Data load

```python
X,y = mglearn.datasets.load_extended_boston()
```

### Data Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Model Building

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train,y_train)
```

```python
print("Training Set Score: {:.2f}".format(ridge.score(X_train,y_train)))
```

```python
print("Test Set Score: {:.2f}".format(ridge.score(X_test,y_test)))
```

At this moment, Model is built with alpha value 1.0(Default Value).

The alpha value is biggeer, Coefficient is closer to zero

Score of Ridge Regression is better, because it can reduce overfitting penalty of modeling

More gernal model

```python
ridge10 = Ridge(alpha=10).fit(X_train,y_train)
```

```python
print("Training Set Score: {:.2f}".format(ridge10.score(X_train,y_train)))
print("Test Set Score: {:.2f}".format(ridge10.score(X_test,y_test)))
```
